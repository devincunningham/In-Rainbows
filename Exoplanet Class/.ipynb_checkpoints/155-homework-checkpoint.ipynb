# 4  Mag. fields

```python
B_J = 2.83 * 10**20 #T·m3
J_radius = 71492000
B_S = 4.6 * 10**18 #T•m3
S_radius = 60330000.0 #m
titan_r = 1221870000.0 #m
europa_r = 671034000.
io_r = 421700000.
Ganymede_r = 14819000000.
Callisto_r = 10759000000.




```

```python
N = 250
dx = S_radius / 2.
Saturn_local = zeros((N,N))
fig=plt.figure(figsize=(10,10))

for j in arange(0,N):
    for i in arange(0,N):
        if ( (j-(N/2))**2+(i-(N/2))**2 ) == 0:
            Saturn_local[j,i] = 2*Saturn_local[j,i-1]
        else: 
            Saturn_local[j,i] = B_S /sqrt( ((j-(N/2))*(dx)   )**2+((i-(N/2))*(dx)   )**2 ) #B_S / ( ((j)*S_radius/100)**2+((i-500)*S_radius/100)**2 )

plt.imshow(np.log10(Saturn_local),cmap= plt.cm.Spectral_r,)
plt.plot(125,125,'g.',ms=12,label='Saturn')
plt.plot(145,125,'r.',ms=2,label='titan')
plt.title('Magnetic field of Saturn')
plt.legend(loc='upper left',numpoints=1,fontsize=14) 
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

Here the steps on the X and Y axis are in terms of Half a Saturn radii.

```python
N = 500
dx = J_radius*2.
Jupiter_local = zeros((N,N))
fig=plt.figure(figsize=(10,10))

for j in arange(0,N):
    for i in arange(0,N):
        if ( (j-(N/2))**2+(i-(N/2))**2 ) == 0:
            Jupiter_local[j,i] = 2*Jupiter_local[j,i-1]
        else: 
            Jupiter_local[j,i] = B_J /sqrt( ((j-(N/2))*(dx)   )**2+((i-(N/2))*(dx)   )**2 ) #B_S / ( ((j)*S_radius/100)**2+((i-500)*S_radius/100)**2 )

plt.imshow(np.log10(Saturn_local),cmap= plt.cm.Spectral_r,)
plt.plot(125,125,'g.',ms=16,label='Jupiter')
plt.plot(130,125,'k.',ms=4,label='europa')
plt.plot(122,125,'r.',ms=4,label='io')
plt.plot(125,125-104,'m.',ms=8,label='Ganymede')
plt.plot(125,125+76,'y.',ms=6,label='Callisto')
legend(loc='upper left',numpoints=1,fontsize=14) 
plt.title('Magnetic field of Jupiter')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

Here the steps on the X and Y axis are in terms of two Jupiter radii.

# 6 Cm for Various systems

```python
def center_of_mass(px,mx):
    cm = sum(px*mx) / sum(mx)
    return cm
Sol_r = 6.95700*10**8
Sol = np.array([1.988e30,0.0])
merc = np.array([0.330e24,(57.9*10**9)])
venus = np.array([4.87e24,(108.9*10**9)])
earth = np.array([5.97e24,(152.1*10**9)])
mars = np.array([0.642e24,(249.2*10**9)])
jupiter = np.array([1898.0e24,(816.6*10**9)])
saturn = np.array([568.0e24,(1514.5*10**9)])
uranus = np.array([86.8e24,(2872.5*10**9)])
neptune = np.array([102e24,(4495.1*10**9)])
pluto = np.array([0.0146e24,(5906.4*10**9)])

# Sun jupiter saturn system
position1 = np.array([Sol[1],jupiter[1],saturn[1]])
mass1 = np.array([Sol[0],jupiter[0],saturn[0]])

#Sol planetary system
position2 = np.array([Sol[1],merc[1],venus[1],earth[1],mars[1],jupiter[1],saturn[1],uranus[1],neptune[1],pluto[1]])
mass2 = np.array([Sol[0],merc[0],venus[0],earth[0],mars[0],jupiter[0],saturn[0],uranus[0],neptune[0],pluto[0]])

# Sun mix jupiter saturn system
position3 = np.array([Sol[1],-jupiter[1],saturn[1]])
mass3 = np.array([Sol[0],jupiter[0],saturn[0]])
            
#Sol mix planetary system
position4 = np.array([Sol[1],-merc[1],venus[1],-earth[1],mars[1],-jupiter[1],saturn[1],-uranus[1],neptune[1],-pluto[1]])
mass4 = np.array([Sol[0],merc[0],venus[0],earth[0],mars[0],jupiter[0],saturn[0],uranus[0],neptune[0],pluto[0]])


oneside_SJS_cm = center_of_mass(position1,mass1)
oneside_All_cm = center_of_mass(position2,mass2)

mixed_SJS_cm = center_of_mass(position3,mass3)
mixed_All_cm = center_of_mass(position4,mass4)

print ('The difference between the two systems is %2.2f percent which is significant as suspected.') %( 100*( abs(oneside_SJS_cm-oneside_All_cm)/oneside_All_cm)) 
print ('The difference between the two systems is: %2.2g m which is %2.2g a solar radius. This is with all planets on one side.') %( abs(oneside_SJS_cm-oneside_All_cm), abs(oneside_SJS_cm-oneside_All_cm)/Sol_r) 
print ('The difference between the two systems is: %2.2g m which is %2.2g a solar radius. This is with mixed location planets.') %( abs(mixed_SJS_cm-mixed_All_cm), abs(mixed_SJS_cm-mixed_All_cm)/Sol_r) 
```

This shows that the wobble of the sun is still within its own radius which explains why the wobble of stars is so hard to detect especially from witin our own solar system.

# 8 human eye resolving power

```python
distance_between_points = .005 #m
D = array([6.,8.25,12.])
theta = arctan(distance_between_points/D) * (360. / (2*pi))
print theta 
print ('The angular resolution we achieved was approx %2.2f arc seconds. ') %(sum(theta*60)/3.)

```

Several factors contributed to our results, including the color of the background and the color of the objects as well as the size of the objects relative to their seperation. The actual resolution of the human eye seems to about half of what we found, 1 arc sec.
