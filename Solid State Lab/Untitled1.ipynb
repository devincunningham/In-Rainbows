```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def E(N,boundary='periodic'):
    S = np.ones(N)
    S[1::2] *= -1
    if boundary == 'open':
        S = S.tolist()
        S.append(0)
        E = (1.0 / N) * np.sum([S[i] * S[i-1] for i, val in enumerate(S)])
    if boundary == 'periodic':
        S = S.tolist()
        S.append(S[0])
        E = (1.0 / N) * np.sum([S[i] * S[i-1] for i, val in enumerate(S)])
    return E

E(100, boundary='periodic')
```

```python
data = [E(i) for i in range(2,100)]
```
