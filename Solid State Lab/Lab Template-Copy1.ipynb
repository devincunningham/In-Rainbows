```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
```

```python
short_counts = np.array([2,5,5,1,4,4,5,4,4,1,2,2,1,2,5,3,2,9,3,1,3,6,3,1,4])
long_counts = np.array([220,169,214,199,174,191,232,223,176,208,223,188,194,218,186,208,192,182,209,212,220,183,211,200])
```

```python
height_ratios = [.05,8]
width_ratios = [8,.05,8]
wspace = 0
hspace = 0
fig = plt.figure(figsize=(sum(width_ratios) + wspace * (len(width_ratios) - 1),
                          sum(height_ratios) + hspace * (len(height_ratios) - 1)))
gs = gridspec.GridSpec(len(height_ratios), len(width_ratios), height_ratios=height_ratios, width_ratios=width_ratios)
fig.suptitle('Cs137 Counts', fontsize=20)

ax = fig.add_subplot(gs[1, 0])
plt.hist(short_counts, bins=15, color='navy', alpha=.75, edgecolor='k')
ax.update({'title': 'Short Counts',
            'xlabel': '',
            'ylabel': 'Number Counts'})
ax.minorticks_on()
ax.tick_params('both', length=8, which='major')
ax.tick_params('both',length=3, which='minor')
ax.grid(True, which='major', ls='dashed', alpha=.5)
ax.grid(True, which='minor', ls='dashed', alpha=.15)
ax.legend(loc='best')

ax = fig.add_subplot(gs[1,2])
plt.hist(long_counts, bins = 15, color='tomato', alpha=.75, edgecolor='k')
ax.update({'title': 'Long Counts',
            'xlabel': '',
            'ylabel': 'Number Counts'})
ax.minorticks_on()
ax.tick_params('both', length=8, which='major')
ax.tick_params('both',length=3, which='minor')
ax.grid(True, which='major', ls='dashed', alpha=.5)
ax.grid(True, which='minor', ls='dashed', alpha=.15)
ax.legend(loc='best')

plt.show()
```

```python
plt.hist(short_counts, bins=15, color='navy', alpha=.75, edgecolor='k')
plt.show()
plt.hist(long_counts, bins = 15, color='tomato', alpha=.75, edgecolor='k')
plt.show()
```
