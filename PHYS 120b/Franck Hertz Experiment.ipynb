### Introduction and theoretical overview:

By means of atomic collision processes, James Franck and Gustav Hertz experimentally demonstrated the existence of quantized energy levels in atoms.$^{[1]}$ This is in agreement with Bohr's model of the atom.

As current of the collector plate tends to increase in relation to the increase in voltage. However, at certain intervals of voltage, the current reaches a local maximum and subsequently drops in magnitude, all before rising again to the next local maximum. The spacing between the maximum points occur at a spacing of 4.9 eV.$^{[2]}$

The number of local maximums coincides with the number of collisions the Mercury vapor has made on its trip to the collector plate. The atoms, which are now excited by the electron collisions, emit photons and drop into their ground state, which uncoincidentally reveals a difference in energy of 4.89eV.

This is in agreeance with both quantum theory and spectrscopic experiments.

### Experimental Setup and Procedures:
To perform this experiment, we use a mercury-filled Franck-Hertz tube, a heat chamber, a control unit for power supply, and a DC current amplifier.

After connecting the BNC socket to our oscilloscope, we powered the heating chamber to 145 degrees Celsius and recorded the current and accelerating voltage data curve with our oscilloscope.

We proceeded to repeat this data acquisition for temperatures of 160 degrees Celsius, 175 degrees Celsius, 190 degrees Celsius, and 210 degrees Celsius.

By recording the local maximums of our data curves, we visualized the average separation and calculated to find our energy spacings.

### Discussion and Conclusion:
We show that atoms can only absorb a discrete amount of energy. Our primary figure shows that anode current vs the accelerating voltage measured from the Franck-Hertz tube at varying temperatures. As one can clearly see, there is a stark trend in the magnitude of anode current and the location of local maximums and minimums for varying temperatures. Visualized in our second figure is the relationship between the minima current and the number of local minima. Further analysis into the average spacing of the minima reveals a value of 4.96 eV. This is a very well measured value in comparison to the expected value of 4.89 eV at a percent difference of 1.43%.
