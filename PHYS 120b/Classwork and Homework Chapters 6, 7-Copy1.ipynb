### 8.8

```python
t = np.array([-4, -2, 0, 2, 4])
s = np.array([13, 25, 34, 42, 56])
delta = np.float((len(t)*sum(t**2) - sum(t)**2))
A = (sum(t**2)*sum(s) - sum(t)*sum(t*s)) / delta
B = (len(t) * sum(t*s) - sum(t)*sum(s)) / delta
m, b = np.polyfit(t, s, 1) 

dt = t[-1] - t[0]
ds = s[-1] - s[0]
print "ds/dt:", ds/dt
print "Manual least squares fit: y = {}x + {}".format(B,A)
print "Numpy linear fit: y = {}x + {}".format(m, b)
plt.scatter(t, s, c='indianred')
plt.plot(t, m*t + b, '--')
plt.plot(t, A + B*t, color='blue')
plt.xlabel('Time (t)')
plt.ylabel('Distance {cm}')
plt.show()
```

### 8.14

```python
sigma_y = np.sqrt((1.0 / (len(t)-2)) * sum((s - A - B*t)**2))
sigma_A = sigma_y * np.sqrt(sum(t**2) / delta)
sigma_B = sigma_y * np.sqrt(len(t)/delta)

print "The student's best estimate for velocity is {} +/- {:.3f} cm/s.".format(B, sigma_B)
plt.scatter(t, s, c='indianred')
plt.plot(t, m*t + b, '--')
plt.plot(t, A + B*t, color='blue')
plt.xlabel('Time (t)')
plt.ylabel('Distance {cm}')
plt.show()
```
