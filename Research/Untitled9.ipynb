```python
UCDs = np.genfromtxt("stephanieradec.csv", skip_header=0, names=True, delimiter=',', skip_footer=0, unpack=True)
print UCDs['ra']
print UCDs['dec']
```

```python
url = 'http://cda.harvard.edu/cscvo/coneSearch?'
i=0
sources = vo.conesearch(url, pos=(UCDs['ra'][i], UCDs['dec'][i]), radius=.008)
print len(sources)
```

```python
print sources[0]
```

```python
for row in sources:
    print ("{0} {1} {2} {3}".format(row["ra"], row["dec"], row["hard_hm"], row["hard_ms"]))
```
