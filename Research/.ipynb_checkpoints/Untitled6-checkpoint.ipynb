```python
fitsRA = 190.53610
fitsDEC = 11.75421
framesize = 5*u.arcmin
frame='icrs'
VLA_filename=None
Chandra_filename=None

myLocation = SkyCoord(fitsRA*u.deg, fitsDEC*u.deg, frame = frame)

VLA_query = vo.sia.SIAQuery('https://skyview.gsfc.nasa.gov/cgi-bin/vo/sia.pl?survey=nvss&',
                            pos=(myLocation.ra.deg, myLocation.dec.deg),
                            size = framesize, format='image/fits',
                            intersect='overlaps')
if VLA_filename is None:
    VLA_filename = 'VLA_RA{}_DEC{}.fits'.format(myLocation.ra.deg, myLocation.dec.deg)
else:
    VLA_filename = VLA_filename
VLA_results = VLA_query.execute()
try:
    VLA_url = VLA_results[0].getdataurl()
    urlretrieve(VLA_url, VLA_filename)
    VLA_fitsimage = '{}'.format(VLA_filename)
except IndexError:
    VLA_fitsimage = 0

Chandra_query = vo.sia.SIAQuery('http://cda.harvard.edu/cxcsiap/queryImages?',
                                pos=(myLocation.ra.deg, myLocation.dec.deg),
                                size = framesize, format='image/fits',
                                intersect='overlaps')
if Chandra_filename is None:
    Chandra_filename = 'Chandra_RA{}_DEC{}.fits'.format(myLocation.ra.deg, myLocation.dec.deg)
else:
    Chandra_filename = Chandra_filename
Chandra_results = Chandra_query.execute()
try:
    Chandra_url = Chandra_results[0].getdataurl()
    urlretrieve(Chandra_url, Chandra_filename)
    Chandra_fitsimage = '{}'.format(Chandra_filename)
except IndexError:
    Chandra_fitsimage = 0

SDSS_fitsimage = SDSS.get_images(coordinates=myLocation)[0]
if SDSS_fitsimage is None:
    SDSS_fitsimage = 0
```

```python
height_ratios = [5]
width_ratios = [5,5,5]
wspace = 0
hspace = 0
fig = plt.figure(figsize=(sum(width_ratios) + wspace *
                          (len(width_ratios) - 1),
                          sum(height_ratios) + hspace *
                          (len(height_ratios) - 1)))
gs = gridspec.GridSpec(len(height_ratios),
                       len(width_ratios),
                       height_ratios=height_ratios,
                       width_ratios=width_ratios)
fig.suptitle('', fontsize=20)

names = ['VLA', 'SDSS', 'Chandra']
for i, image in enumerate([VLA_fitsimage, SDSS_fitsimage, Chandra_fitsimage]):
    ax = fig.add_subplot(gs[i])
    ax.set_xlabel('')
    ax.set_ylabel('')
    ax.xaxis.set_ticks_position('none')
    ax.yaxis.set_ticks_position('none')
    ax.set_xticklabels([])
    ax.set_yticklabels([])
    if image != 0:
        ax = aplpy.FITSFigure(image,figure=fig,
                              subplot=list(gs[i].get_position(fig).bounds))
        ax.show_grayscale(stretch='log', vmin=1e-5)
        ax.show_markers(myLocation.ra,myLocation.dec)
#         ax.show_contour()
        ax.recenter(myLocation.ra,myLocation.dec,radius=.0333)
        ax.add_scalebar(1*u.arcminute, '1 Arcminute', color='red')
        ax.set_title('{}'.format(names[i]))
        ax.set_xaxis_coord_type('scalar')
        ax.set_yaxis_coord_type('scalar')
        ax.tick_labels.set_xformat('%11.3f')
        ax.tick_labels.set_yformat('%11.3f')
    elif image == 0:
        ax = plt.text(0.5, 0.5,'{} Image Not Found'.format(names[i]),
                      horizontalalignment='center',
                      verticalalignment='center',
                      transform = ax.transAxes)

plt.show()
```
